**Using XGBoost Regressor**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
import xgboost
```

Importing California House Price Data

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# loading dataset to panda dataframe
df = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
df.head()
```

```python
# Add the target columns to the dataframe
df['price'] = house_price_dataset.target
```

```python
df.head()
```

```python
#checking no of rows and columns of dataset
df.shape
```

```python
#Check for missing values
df.isnull().sum()
```

```python
# statistical measures
df.describe()
```

```python
#Understanding the Correlation between various features of dataset
#Positive Correlation
#Negative Correlation
```

```python
correlation = df.corr()
print (correlation)
```

```python
# consruction a heat map to understand the correlation
plt.figure(figsize=(8,8))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot= True, annot_kws={'size' : 8}, cmap = 'Blues')
```

```python
#splitting the data into x and y
X = df.drop(['price'], axis = 1)
y = df['price']
```

```python
print(X,y)
```

```python
X.shape
```

```python
#Splitting the data into training and test data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
#XGBoost Regressor
model = XGBRegressor()
model.fit(x_train,y_train, eval_set = [(x_train,y_train)])
```

R square error

```python
x_test_pred = model.predict(x_test)
accuracy_test = metrics.r2_score(y_test, x_test_pred)
print(accuracy_test)
```

```python
x_train_pred = model.predict(x_train)
accuracy_train = metrics.r2_score(y_train, x_train_pred)
print(accuracy_train)
```

Mean Absoute error

```python
print(metrics.mean_absolute_error(y_test, x_test_pred))
```
